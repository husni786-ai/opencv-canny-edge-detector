```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Step 1: Read the image
image = cv2.imread("/content/cat.jpeg")

# Step 2: Apply filters
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(image, (25, 25), 0)
edges = cv2.Canny(image, 100, 200)   #canny = edge detection algo

kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
sharpen = cv2.filter2D(image, -1, kernel)

# Step 3: Show results using cv2_imshow()
cv2_imshow(image)
cv2_imshow(gray)
cv2_imshow(blur)
cv2_imshow(edges)
cv2_imshow(sharpen)

# Step 4: Release windows (not needed in Colab)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Step 1: Read the image
image = cv2.imread("/content/Dog.jpeg")

# Step 2: Check if the image loaded successfully
if image is None:
    print("âš  Image not found! Please upload again.")
else:
    # Step 3: Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Step 4: Apply Gaussian blur to reduce noise
    blur = cv2.GaussianBlur(gray, (5, 5), 0)

    # Step 5: Apply Canny Edge Detection
    edges = cv2.Canny(blur, 100, 200)

    # Step 6: Show results
    print("âœ… Original Image:")
    cv2_imshow(image)

    print("âœ… Grayscale Image:")
    cv2_imshow(gray)

    print("âœ… Blurred Image:")
    cv2_imshow(blur)

    print("âœ… Edge Detected Image:")
    cv2_imshow(edges)
```
